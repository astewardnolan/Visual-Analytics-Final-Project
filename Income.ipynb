# Line Graphs for Income by City

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

## Seattle:

```python
# Reading in the Seattle csv
seattle = pd.read_csv("SEAT653PCPI.csv")
```

```python
# Standardizing date format
seattle['DATE'] = pd.to_datetime(seattle['DATE'])

# Filtering the csv for the years 2013-2023
filtered_seattle = seattle[(seattle['DATE'].dt.year >= 2013) & (seattle['DATE'].dt.year <= 2023)]
filtered_seattle
```

```python
# Plotting our data in a line graph
plt.plot(filtered_seattle['SEAT653PCPI'], filtered_seattle['DATE'], marker='o', color='#40c9cd')
plt.xlabel('Per-Capita Income')
plt.ylabel('Year')
plt.title("Seattle's Per-Capita Income by Year")
plt.show()
```
