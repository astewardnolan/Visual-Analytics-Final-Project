# Line Graphs for Median Sales Price by City

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading in the data
df = pd.read_csv('MedianSalesPricesforCities.csv')
df
```

```python
# Dropping unneccessary columns
df = df.drop(columns=['Unnamed: 4', 'Unnamed: 5'])
df
```

```python
# df = df.set_index('Metropolitan Area')
# df
```

## Seattle:

```python
# Extract the row where the 'Metropolitan' row equals 'Seattle'
seattle = df.loc[df['Metropolitan Area'] == 'Seattle-Tacoma-Bellevue, WA']
print(seattle)
```

```python
x_seattle = seattle.columns[1:]  # Years (ignores 'Metropolitan Area')
y_seattle = seattle.iloc[0, 1:]  # Values (row values from 2021 onwards)
print(x_seattle)
print(y_seattle)
```

```python
# # Plotting the seattle data from the df
# # type(row)
# plt.plot(seattle['Metropolitan Area'], seattle['DATE'], marker='o', color='#40c9cd')
# plt.xlabel('Median Sales Price')
# plt.ylabel('Year')
# plt.title("Median Sales Price of a Home in Seattle by Year")
# plt.show()
```

```python
plt.plot(x_seattle, y_seattle, marker='o', color='#40c9cd')
plt.title('Median Sales Price for Seattle-Tacoma-Bellevue, WA')
plt.xlabel('Year')
plt.ylabel('Median Sales Price (in Thousands)')
# plt.grid(True)
plt.show()
```
